<h4>1. Cost Function</h4>

What is the best line from the data above?

We want to calculate values of $m$ and $c$ that give the lowest value for the cost value above.

For our given data set we can plot the cost value/function.

Recall that the cost is:

$$ Cost(m, c) = \sum_i (y_i - mx_i - c)^2 $$

````python
# Calculate the cost of the lines above for the data above.
cost = lambda m,c: np.sum([(y[i] - m * x[i] - c)**2 for i in range(x.size)])

print("Cost with m = %5.2f and c = %5.2f: %8.2f" % (0.1, 1.2, cost(0.1, 1.2)))
print("Cost with m = %5.2f and c = %5.2f: %8.2f" % (0.09,  1.0, cost(0.09,  1.0)))
print("Cost with m = %5.2f and c = %5.2f: %8.2f" % (0.07, 1.3, cost(0.07, 1.3)))
````

Recreate the plot with the *best* line by calculate the best values for m and c.

````python
# Calculate the best values for m and c.

# First calculate the means (a.k.a. averages) of w and d.
w_avg = np.mean(x)
d_avg = np.mean(y)

# Subtract means from w and d.
w_zero = x - w_avg
d_zero = y - d_avg

# The best m is found by the following calculation.
m = np.sum(w_zero * d_zero) / np.sum(w_zero * w_zero)
# Use m from above to calculate the best c.
c = d_avg - m * w_avg

print("m is %8.6f and c is %6.6f." % (m, c))
````

We can also use https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html to output the best values for m and c. The Polynomial.fit class method is recommended for new code as it is more stable numerically

````python
# Calculate the cost of the lines above for the data above using numpy polyfit
np.polyfit(x, y, 1)
````

````python
plt.plot(x, y, 'k.')

# Calculate the best values for m and c.

plt.plot(x, 0.10315789 * x + 1.04947368, color="red", label=r"$0.10315789x + 1.049474$")

# Add a legend.
plt.legend()

# Add axis labels.
plt.xlabel('Point from x')
plt.ylabel('Point from y')
plt.title('the best values for m and c.')

# Show the plot.
plt.show()
````

<h3>2. Scipy</h3>

Using scipy to estimate the parameters.

$$ Cost(m, c) = \sum_i (y_i - mx_i - c)^2 $$

````python
# Optimization.
import scipy.optimize as so

# Numerical arrays.
import numpy as np
````

````python
# We are going to use minimize function in scipy.optimize.
# So - what do we need? First we need a function to minimze.

# x and y are have fixed values in this scenario as above.

# The function to be minimized - the cost function.
# It must take a single argument! But our cost function takes 2: m and c.
# So, we'll put them in a list called MC. e.g. MC = (5, 10)
def cost(MC):
    # Unpack the values m and c.
    m, c = MC
    # We need the data points in this function.
    cost = np.sum((y - m * x - c)**2)
    # Return that value.
    return cost
````

````python
# Cost with m =  0.10 and c =  1.20:     1.30
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize
result = so.minimize(cost, (0.10, 1.20))
# Show the results.
result
````

````python
# Extract the optimized m and c.
m_o, c_o = result.x
# Print them.
m_o, c_o
````

````python
# Previous, analytical result.
m_a, c_a = np.polyfit(x, y, 1)
# Print analytical results.
m_a, c_a
````

````python
# Plot the best fit line from the optimization.
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'k.', label='Original data')
ax.plot(x, m_o * x + c_o, 'b-', label='Optimized line')
ax.plot(x, m_a * x + c_a, 'g-', label='Analytical line')
ax.legend()

# Show the plot.
plt.show()
````

<h3>3. Curve fitting</h3>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

````python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html
# Create the model function.
def f(x, m, c):
    return m * x + c
````

````python
# Run curve_fit.
result = so.curve_fit(f, x, y)
````

````python
# Look at the result.
result
````

````python
# Pull out the parameters.
m_f, c_f = result[0]
# Print values.
m_f, c_f
````

````python
# Plot the best fit line from the optimization.
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'k.', label='Original data')
ax.plot(x, m_f * x + c_f, 'r-', label='Curve Fit line')
ax.plot(x, m_a * x + c_a, 'g-', label='Analytical line')
ax.legend()

# Show the plot.
plt.show()
````
